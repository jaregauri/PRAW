```python
#saving the scrapped data into csv file

import pandas as pd
date = []
post_score = []
post_title = []
top_comment_1_list = []
top_comment_2_list = []
top_comment_3_list = []
#top posts from all time
for submission in reddit.subreddit("datascience").top(limit = 200):
    
        
    temp = submission.title.split()
    if len(temp) > 5:
        date_value = str(time.gmtime(submission.created_utc).tm_year) + '-' + str(time.gmtime(submission.created_utc).tm_mon) + '-' + str(time.gmtime(submission.created_utc).tm_mday)
        date.append(date_value)
        post_score.append(submission.score)
        post_title.append(submission.title)

        submissionObject = reddit.submission(id=submission)
        submissionObject.comments.replace_more(limit=None)
        for top_level_comment in submissionObject.comments:
            comment_list.append(top_level_comment.body)
        x = 0   
        if (len(comment_list)) > 3:
            while x < 3:
                if x == 0 :
                    top_comment_1_list.append(comment_list[x])
                elif x == 1:
                    top_comment_2_list.append(comment_list[x])
                elif x == 2:
                    top_comment_3_list.append(comment_list[x])
                x = x + 1

        comment_list = []     
#         y = y - 1
    data = { 'Date' : date, 'Post Score' : post_score, 'Post_title' : post_title , 'Top Comment 1' :  top_comment_1_list , 
            'Top Comment 2' : top_comment_2_list ,  'Top_Comment_3' : top_comment_3_list}

pd.set_option("display.max_rows", None, "display.max_columns", None)

df = pd.DataFrame(data=data)
df
```

```python
#extracting 101 rows from the dataframe and saving it to reddit_data.csv
df = df[0:101]
df.to_csv('reddit_data.csv', index = False)
```
